```python
dataset = dataset.loc[dataset['article_nom'] == 'Blanc De Poulet 2T 80G Casino']
```

```python
dataset = dataset.drop(columns = ['id','date','article_nom','id_categorie_6','id_categorie_5','id_categorie_4','cat6_nom','cat5_nom','cat4_nom'])
```

```python
dataset.head()
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
encode_cities = LabelEncoder()
dataset.implant = encode_cities.fit_transform(dataset.implant)
```

```python
X = dataset.drop(columns = ['qte_article_vendue'])
y = dataset.qte_article_vendue
```

```python
X = np.array(X)
for i in range(X.shape[0]):
    for j in range(X.shape[1]):
        try:
            X[i,j] = '.'.join(X[i,j].split(','))
        except:
            pass
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
encode_qt = LabelEncoder()
y_categories = encode_qt.fit_transform(y)
```

```python
y = np.array(y)
for i in range(len(y)):
    try:
        y[i] = '.'.join(y[i].split(','))
    except:
        pass
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators=50,verbose = 0,n_jobs = 5)
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = { 
            "n_estimators"      : [5000],
            'class_weight': ['balanced',None]
            }

grid_search = GridSearchCV(estimator = classifier,
                           param_grid = param_grid,
                           scoring = 'accuracy',
                           cv = 5,
                           n_jobs = 3,
                           verbose = 10)
grid_search = grid_search.fit(X, y_categories)
```
